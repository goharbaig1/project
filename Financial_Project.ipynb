**LSTM Barclays**




```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Step 1: Load the Data (covering the entire dataset)
barc_df = pd.read_csv('/content/drive/My Drive/Training/BARC.L.csv')
print(barc_df.head())

# Step 2: Data Preprocessing
barc_df['Daily Return'] = barc_df['Adj Close'].pct_change()
barc_df = barc_df.dropna()
print(barc_df[['Date', 'Adj Close', 'Daily Return']].head())

plt.figure(figsize=(12, 6))
plt.plot(barc_df['Date'], barc_df['Daily Return'], label='Daily Return')
plt.title('Barclays (BARC.L) - Daily Returns')
plt.xlabel('Date')
plt.ylabel('Return')
plt.legend()
plt.show()

# Step 3: Prepare the Data for LSTM
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(barc_df['Adj Close'].values.reshape(-1,1))

# Define the window size based on train_data length
train_size = int(len(scaled_data) * 0.8)
window_size = min(1800, train_size - 1)  # Ensure the window size allows for some training samples
train_data, test_data = scaled_data[0:train_size,:], scaled_data[train_size-window_size:len(scaled_data),:]

# Debugging: Check the lengths before entering the loop
print(f"Length of train_data: {len(train_data)}")
print(f"Adjusted window size: {window_size}")

# Prepare training data using the adjusted window size
X_train, y_train = [], []
for i in range(window_size, len(train_data)):
    X_train.append(train_data[i-window_size:i, 0])
    y_train.append(train_data[i, 0])

X_train, y_train = np.array(X_train), np.array(y_train)

# Debugging Step: Check the shape before reshaping
print(f"Shape of X_train before reshaping: {X_train.shape}")

# Reshape X_train to be 3D for LSTM [samples, timesteps, features]
if X_train.shape[0] > 0 and X_train.shape[1] > 0:  # Ensure there is data to reshape
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
else:
    print("Error: X_train is empty or has unexpected shape.")

# Step 4: Build and Train the LSTM Model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(units=1))

model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train, y_train, epochs=25, batch_size=32)

# Save the trained model
model_save_path = '/content/drive/My Drive/Models/LSTM_BARC.h5'
model.save(model_save_path)
print(f'Model saved to {model_save_path}')

# Step 5: Predict on the Test Data
X_test, y_test = [], []
for i in range(window_size, len(test_data)):
    X_test.append(test_data[i-window_size:i, 0])
    y_test.append(test_data[i, 0])

X_test, y_test = np.array(X_test), np.array(y_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

predicted_test_prices = model.predict(X_test)
predicted_test_prices = scaler.inverse_transform(predicted_test_prices)
y_test = scaler.inverse_transform(y_test.reshape(-1, 1))

# Step 6: Predict Future Prices
future_days = 30
# Continue prediction from the last sequence of the test set
test_data_scaled = test_data[-window_size:]  # Last 2190 days from test data

predicted_future_prices = []
for _ in range(future_days):
    X_test_future = np.array([test_data_scaled])
    X_test_future = np.reshape(X_test_future, (X_test_future.shape[0], X_test_future.shape[1], 1))

    predicted_price_future = model.predict(X_test_future)
    predicted_future_prices.append(predicted_price_future[0][0])

    # Update the test_data array to include this predicted price
    test_data_scaled = np.append(test_data_scaled, predicted_price_future)[-window_size:].reshape(-1, 1)

predicted_future_prices = scaler.inverse_transform(np.array(predicted_future_prices).reshape(-1,1))

# Step 7: Evaluate the Model on Test Data
mse_lstm_test = mean_squared_error(y_test, predicted_test_prices)
r2_lstm_test = r2_score(y_test, predicted_test_prices)

print(f'LSTM MSE on Test Data: {mse_lstm_test}')
print(f'LSTM R² on Test Data: {r2_lstm_test}')

# Step 8: Plotting the Predictions
plt.figure(figsize=(12, 6))
plt.plot(range(len(y_test)), y_test, color='blue', label='Actual Barclays Stock Price')
plt.plot(range(len(predicted_test_prices)), predicted_test_prices, color='red', label='Predicted Stock Prices (Test)')
plt.plot(range(len(y_test), len(y_test) + future_days), predicted_future_prices, color='green', label='Predicted Future Stock Prices')
plt.title('Barclays (BARC.L) Stock Price Prediction')
plt.xlabel('Days')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import joblib

# Load the data
barc_df = pd.read_csv('/content/drive/My Drive/Training/BARC.L.csv')
barc_df = barc_df.dropna(subset=['Adj Close'])

# Define the window size based on train_data length
train_size = int(len(barc_df) * 0.8)
window_size = min(1800, train_size - 1)  # Ensure the window size allows for some training samples

# Debugging: Print the window size and train_size
print(f"Window size: {window_size}, Train size: {train_size}")

# Creating the feature and target datasets
X_rf, y_rf = [], []

# Training data preparation
for i in range(window_size, train_size):
    X_rf.append(barc_df['Adj Close'].iloc[i-window_size:i].values)
    y_rf.append(barc_df['Adj Close'].iloc[i])

X_train_rf = np.array(X_rf)
y_train_rf = np.array(y_rf)

# Testing data preparation (starts from train_size - window_size)
X_rf, y_rf = [], []

for i in range(train_size, len(barc_df)):
    X_rf.append(barc_df['Adj Close'].iloc[i-window_size:i].values)
    y_rf.append(barc_df['Adj Close'].iloc[i])

X_test_rf = np.array(X_rf)
y_test_rf = np.array(y_rf)

# Ensure that the lengths are consistent
print(f"Length of X_train_rf: {len(X_train_rf)}, y_train_rf: {len(y_train_rf)}")
print(f"Length of X_test_rf: {len(X_test_rf)}, y_test_rf: {len(y_test_rf)}")

# Step 2: Train the Random Forest Regressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train_rf, y_train_rf)

# Save the trained model
model_save_path = '/content/drive/My Drive/Models/RF_BARC_model.pkl'
joblib.dump(rf_model, model_save_path)
print(f'Model saved to {model_save_path}')

# Step 3: Predict Future Prices
# Start with the last sequence from the test set
last_sequence = X_test_rf[-1].reshape(1, -1)

# Initialize an array to store predicted future prices
future_prices = []

# Iteratively predict the next 30 days
for _ in range(30):
    predicted_price = rf_model.predict(last_sequence)
    future_prices.append(predicted_price[0])

    # Update the last sequence by dropping the first value and appending the new predicted price
    last_sequence = np.append(last_sequence[:, 1:], predicted_price).reshape(1, -1)

# Step 4: Make Predictions on the Test Data
y_pred_rf = rf_model.predict(X_test_rf)

# Step 5: Plot the results
plt.figure(figsize=(12, 6))
plt.plot(range(len(y_test_rf)), y_test_rf, color='blue', label='Actual Prices')  # Plot actual prices from test data
plt.plot(range(len(y_pred_rf)), y_pred_rf, color='red', label='Predicted Prices')  # Plot predicted prices on test data
plt.plot(range(len(y_test_rf), len(y_test_rf) + 30), future_prices, color='green', label='Future Predicted Prices')  # Plot future predicted prices
plt.title('Barclays (BARC.L) Stock Price Prediction with Random Forest')
plt.xlabel('Days')
plt.ylabel('Stock Price')
plt.legend()
plt.show()

# Print the predicted stock prices for the next 30 days
print("Predicted Stock Prices for the next 30 days:")
print(future_prices)

# Step 6: Evaluate the Model on Test Data (before future prediction)
mse_rf = mean_squared_error(y_test_rf, y_pred_rf)
r2_rf = r2_score(y_test_rf, y_pred_rf)
print(f'Random Forest MSE on Test Data: {mse_rf}')
print(f'Random Forest R² on Test Data: {r2_rf}')
```

**LSTM Lloyds**

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Step 1: Load the Data (covering the entire dataset)
lloy_df = pd.read_csv('/content/drive/My Drive/Training/LLOY.L.csv')
print(lloy_df.head())

# Step 2: Data Preprocessing
lloy_df['Daily Return'] = lloy_df['Adj Close'].pct_change()
lloy_df = lloy_df.dropna()
print(lloy_df[['Date', 'Adj Close', 'Daily Return']].head())

plt.figure(figsize=(12, 6))
plt.plot(lloy_df['Date'], lloy_df['Daily Return'], label='Daily Return')
plt.title('Lloyds Banking Group (LLOY.L) - Daily Returns')
plt.xlabel('Date')
plt.ylabel('Return')
plt.legend()
plt.show()

# Step 3: Prepare the Data for LSTM
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(lloy_df['Adj Close'].values.reshape(-1,1))

# Define the window size based on train_data length
train_size = int(len(scaled_data) * 0.8)
window_size = min(1800, train_size - 1)  # Ensure the window size allows for some training samples
train_data, test_data = scaled_data[0:train_size,:], scaled_data[train_size-window_size:len(scaled_data),:]

# Debugging: Check the lengths before entering the loop
print(f"Length of train_data: {len(train_data)}")
print(f"Adjusted window size: {window_size}")

# Prepare training data using the adjusted window size
X_train, y_train = [], []
for i in range(window_size, len(train_data)):
    X_train.append(train_data[i-window_size:i, 0])
    y_train.append(train_data[i, 0])

X_train, y_train = np.array(X_train), np.array(y_train)

# Debugging Step: Check the shape before reshaping
print(f"Shape of X_train before reshaping: {X_train.shape}")

# Reshape X_train to be 3D for LSTM [samples, timesteps, features]
if X_train.shape[0] > 0 and X_train.shape[1] > 0:  # Ensure there is data to reshape
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
else:
    print("Error: X_train is empty or has unexpected shape.")

# Step 4: Build and Train the LSTM Model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(units=1))

model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train, y_train, epochs=25, batch_size=32)

# Save the trained model
model_save_path = '/content/drive/My Drive/Models/LSTM_LLOY_model.h5'
model.save(model_save_path)
print(f'Model saved to {model_save_path}')

# Step 5: Predict on the Test Data
X_test, y_test = [], []
for i in range(window_size, len(test_data)):
    X_test.append(test_data[i-window_size:i, 0])
    y_test.append(test_data[i, 0])

X_test, y_test = np.array(X_test), np.array(y_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

predicted_test_prices = model.predict(X_test)
predicted_test_prices = scaler.inverse_transform(predicted_test_prices)
y_test = scaler.inverse_transform(y_test.reshape(-1, 1))

# Step 6: Predict Future Prices
future_days = 30
# Continue prediction from the last sequence of the test set
test_data_scaled = test_data[-window_size:]  # Last window_size days from test data

predicted_future_prices = []
for _ in range(future_days):
    X_test_future = np.array([test_data_scaled])
    X_test_future = np.reshape(X_test_future, (X_test_future.shape[0], X_test_future.shape[1], 1))

    predicted_price_future = model.predict(X_test_future)
    predicted_future_prices.append(predicted_price_future[0][0])

    # Update the test_data array to include this predicted price
    test_data_scaled = np.append(test_data_scaled, predicted_price_future)[-window_size:].reshape(-1, 1)

predicted_future_prices = scaler.inverse_transform(np.array(predicted_future_prices).reshape(-1,1))

# Step 7: Evaluate the Model on Test Data
mse_lstm_test = mean_squared_error(y_test, predicted_test_prices)
r2_lstm_test = r2_score(y_test, predicted_test_prices)

print(f'LSTM MSE on Test Data: {mse_lstm_test}')
print(f'LSTM R² on Test Data: {r2_lstm_test}')

# Step 8: Plotting the Predictions
plt.figure(figsize=(12, 6))
plt.plot(range(len(y_test)), y_test, color='blue', label='Actual Lloyds Banking Group Stock Price')
plt.plot(range(len(predicted_test_prices)), predicted_test_prices, color='red', label='Predicted Stock Prices (Test)')
plt.plot(range(len(y_test), len(y_test) + future_days), predicted_future_prices, color='green', label='Predicted Future Stock Prices')
plt.title('Lloyds Banking Group (LLOY.L) Stock Price Prediction')
plt.xlabel('Days')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import joblib

# Step 1: Load the Data (covering the entire dataset)
vod_df = pd.read_csv('/content/drive/My Drive/Training/VOD.L.csv')
print(vod_df.head())

# Step 2: Data Preprocessing
vod_df['Daily Return'] = vod_df['Adj Close'].pct_change()
vod_df = vod_df.dropna()
print(vod_df[['Date', 'Adj Close', 'Daily Return']].head())

plt.figure(figsize=(12, 6))
plt.plot(vod_df['Date'], vod_df['Daily Return'], label='Daily Return')
plt.title('Vodafone Group (VOD.L) - Daily Returns')
plt.xlabel('Date')
plt.ylabel('Return')
plt.legend()
plt.show()

# Step 3: Prepare the Data for LSTM
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(vod_df['Adj Close'].values.reshape(-1,1))

# Define the window size based on train_data length
train_size = int(len(scaled_data) * 0.8)
window_size = min(1800, train_size - 1)  # Ensure the window size allows for some training samples
train_data, test_data = scaled_data[0:train_size,:], scaled_data[train_size-window_size:len(scaled_data),:]

# Debugging: Check the lengths before entering the loop
print(f"Length of train_data: {len(train_data)}")
print(f"Adjusted window size: {window_size}")

# Prepare training data using the adjusted window size
X_train, y_train = [], []
for i in range(window_size, len(train_data)):
    X_train.append(train_data[i-window_size:i, 0])
    y_train.append(train_data[i, 0])

X_train, y_train = np.array(X_train), np.array(y_train)

# Debugging Step: Check the shape before reshaping
print(f"Shape of X_train before reshaping: {X_train.shape}")

# Reshape X_train to be 3D for LSTM [samples, timesteps, features]
if X_train.shape[0] > 0 and X_train.shape[1] > 0:  # Ensure there is data to reshape
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
else:
    print("Error: X_train is empty or has unexpected shape.")

# Step 4: Build and Train the LSTM Model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(units=1))

model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train, y_train, epochs=25, batch_size=32)

# Step 5: Save the Model
model_save_path = '/content/drive/My Drive/Models/LSTM_VOD.h5'
model.save(model_save_path)
print(f'Model saved to {model_save_path}')

# Step 6: Predict on the Test Data
X_test, y_test = [], []
for i in range(window_size, len(test_data)):
    X_test.append(test_data[i-window_size:i, 0])
    y_test.append(test_data[i, 0])

X_test, y_test = np.array(X_test), np.array(y_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

predicted_test_prices = model.predict(X_test)
predicted_test_prices = scaler.inverse_transform(predicted_test_prices)
y_test = scaler.inverse_transform(y_test.reshape(-1, 1))

# Step 7: Predict Future Prices
future_days = 30
# Continue prediction from the last sequence of the test set
test_data_scaled = test_data[-window_size:]  # Last window_size days from test data

predicted_future_prices = []
for _ in range(future_days):
    X_test_future = np.array([test_data_scaled])
    X_test_future = np.reshape(X_test_future, (X_test_future.shape[0], X_test_future.shape[1], 1))

    predicted_price_future = model.predict(X_test_future)
    predicted_future_prices.append(predicted_price_future[0][0])

    # Update the test_data array to include this predicted price
    test_data_scaled = np.append(test_data_scaled, predicted_price_future)[-window_size:].reshape(-1, 1)

predicted_future_prices = scaler.inverse_transform(np.array(predicted_future_prices).reshape(-1,1))

# Step 8: Evaluate the Model on Test Data
mse_lstm_test = mean_squared_error(y_test, predicted_test_prices)
r2_lstm_test = r2_score(y_test, predicted_test_prices)

print(f'LSTM MSE on Test Data: {mse_lstm_test}')
print(f'LSTM R² on Test Data: {r2_lstm_test}')

# Step 9: Plotting the Predictions
plt.figure(figsize=(12, 6))
plt.plot(range(len(y_test)), y_test, color='blue', label='Actual Vodafone Group Stock Price')
plt.plot(range(len(predicted_test_prices)), predicted_test_prices, color='red', label='Predicted Stock Prices (Test)')
plt.plot(range(len(y_test), len(y_test) + future_days), predicted_future_prices, color='green', label='Predicted Future Stock Prices')
plt.title('Vodafone Group (VOD.L) Stock Price Prediction')
plt.xlabel('Days')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from tensorflow.keras.models import load_model
import joblib

# Step 1: Load and preprocess the data
bats_df = pd.read_csv('/content/drive/My Drive/Testing/BATS.L.csv')
bats_df = bats_df.dropna(subset=['Adj Close'])

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(bats_df['Adj Close'].values.reshape(-1, 1))

# Define the window size based on train_data length
window_size = 1800  # Or use a different size based on the data characteristics

# Prepare the data for testing
X_test = []
for i in range(window_size, len(scaled_data)):
    X_test.append(scaled_data[i-window_size:i, 0])

X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

# Step 2: Load the pre-trained models for all four stocks

# LSTM Models
lstm_model_barc = load_model('/content/drive/My Drive/Models/LSTM_BARC.h5')
lstm_model_lloy = load_model('/content/drive/My Drive/Models/LSTM_LLOY_model.h5')
lstm_model_rio = load_model('/content/drive/My Drive/Models/LSTM_RIO.h5')
lstm_model_vod = load_model('/content/drive/My Drive/Models/LSTM_VOD.h5')

# Random Forest Models
rf_model_barc = joblib.load('/content/drive/My Drive/Models/RF_BARC_model.pkl')
rf_model_lloy = joblib.load('/content/drive/My Drive/Models/RF_LLOY_model.pkl')
rf_model_rio = joblib.load('/content/drive/My Drive/Models/RF_RIO_model.pkl')
rf_model_vod = joblib.load('/content/drive/My Drive/Models/RF_VOD_model.pkl')

# GBM Models
gbm_model_barc = joblib.load('/content/drive/My Drive/Models/GBM_BARC_model.pkl')
gbm_model_lloy = joblib.load('/content/drive/My Drive/Models/GBM_LLOY_model.pkl')
gbm_model_rio = joblib.load('/content/drive/My Drive/Models/GBM_RIO_model.pkl')
gbm_model_vod = joblib.load('/content/drive/My Drive/Models/GBM_VOD_model.pkl')

# Step 3: Make predictions using all models

# LSTM Predictions
lstm_pred_barc = lstm_model_barc.predict(X_test)
lstm_pred_lloy = lstm_model_lloy.predict(X_test)
lstm_pred_rio = lstm_model_rio.predict(X_test)
lstm_pred_vod = lstm_model_vod.predict(X_test)

# Inverse transform LSTM predictions
lstm_pred_barc = scaler.inverse_transform(lstm_pred_barc)
lstm_pred_lloy = scaler.inverse_transform(lstm_pred_lloy)
lstm_pred_rio = scaler.inverse_transform(lstm_pred_rio)
lstm_pred_vod = scaler.inverse_transform(lstm_pred_vod)

# Random Forest Predictions
X_test_rf = X_test.reshape(X_test.shape[0], X_test.shape[1])

rf_pred_barc = rf_model_barc.predict(X_test_rf).reshape(-1, 1)
rf_pred_lloy = rf_model_lloy.predict(X_test_rf).reshape(-1, 1)
rf_pred_rio = rf_model_rio.predict(X_test_rf).reshape(-1, 1)
rf_pred_vod = rf_model_vod.predict(X_test_rf).reshape(-1, 1)

# GBM Predictions
gbm_pred_barc = gbm_model_barc.predict(X_test_rf).reshape(-1, 1)
gbm_pred_lloy = gbm_model_lloy.predict(X_test_rf).reshape(-1, 1)
gbm_pred_rio = gbm_model_rio.predict(X_test_rf).reshape(-1, 1)
gbm_pred_vod = gbm_model_vod.predict(X_test_rf).reshape(-1, 1)

# Step 4: Ensemble averaging

# Combine all predictions
ensemble_pred = (
    lstm_pred_barc + lstm_pred_lloy + lstm_pred_rio + lstm_pred_vod +
    rf_pred_barc + rf_pred_lloy + rf_pred_rio + rf_pred_vod +
    gbm_pred_barc + gbm_pred_lloy + gbm_pred_rio + gbm_pred_vod
) / 12  # Averaging over 12 models

# Step 5: Evaluate the ensemble model
actual_prices = bats_df['Adj Close'].values[window_size:]
mse_ensemble = mean_squared_error(actual_prices, ensemble_pred)
r2_ensemble = r2_score(actual_prices, ensemble_pred)

print(f'Ensemble MSE: {mse_ensemble}')
print(f'Ensemble R²: {r2_ensemble}')

# Step 6: Plot the results
plt.figure(figsize=(12, 6))
plt.plot(actual_prices, color='blue', label='Actual Prices')
plt.plot(ensemble_pred, color='red', label='Ensemble Predicted Prices')
plt.title('BATS.L Stock Price Prediction with Ensemble Averaging')
plt.xlabel('Days')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```
